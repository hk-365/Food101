```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def apply_kernel(image, kernel):
    return cv2.filter2D(image, -1, kernel)

def main():
    image_path = input("Enter the path of the image: ")

    image = cv2.imread(image_path)

    if image is None:
        print("Error: Could not load the image. Please check the path and try again.")
        return None
    return image

sharpen_kernel = np.array([[-1, -1, -1],
                            [-1, 9, -1],
                            [-1, -1, -1]])

blur_kernel = np.ones((6, 6), dtype=np.float32) / 36

edge_kernel = np.array([[-1, -1, -1],
                        [-1, 8, -1],
                        [-1, -1, -1]])
```

```python
image=main()
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
sharp_image = apply_kernel(image, sharpen_kernel)
blur_image = apply_kernel(image, blur_kernel)
edge_image = apply_kernel(grayscale_image, edge_kernel)
```

```python
def show_img(image):
    image_rgb = cv2.cvtColor(image.astype(np.uint8), cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.show()
```

```python
def show_img_gray(image):
    image_rgb = cv2.cvtColor(image.astype(np.uint8), cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb,cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
print("image")
show_img(image)
print("sharp_image")
show_img(sharp_image)
print("blur_image")
show_img(blur_image)
print("edge_image")
show_img_gray(edge_image)
```

```python
cv2.imwrite("sharpened_image.jpg", sharp_image)
cv2.imwrite("blurred_image.jpg", blur_image)
cv2.imwrite("edge_highlighted_image.jpg", edge_image)
print("Images have been processed and saved as:")
print("1. sharpened_image.jpg")
print("2. blurred_image.jpg")
print("3. edge_highlighted_image.jpg")
```
